The scatter matrix allows us to get a high level overview of both relationships and distributions of our data. Now lets use scatter plots to explore these interesting variables further.

```python
plt.scatter(df.LSTAT, df.MEDV)
plt.xlabel("Lower Status of Population");
plt.ylabel("Median Value of Home");
plt.title("LSTAT vs MEDV");
```

One thing that should raise a red flag here is that cluster in a straight line at the top of the chart on the y-axis. Let's explore this further.

```python
display(df[df.MEDV == 50])
```

It appears as if 50 (or $500,000) is the cap for this dataset. Therefore all houses that are above this limit are automatically set as 50. In practice this would be a situation where you must discuss this issue with the dataset creator or another colleague with the specific domain knowledge relevant to the problem at hand. In this case, I will choose to drop these values. Let's quickly train a model to show case the effect outliers can have on a model.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.2, random_state=23)
```

```python
x_train.shape, x_test.shape
```

```python
 y_train.shape, y_test.shape
```

```python
y_train = y_train.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
original_preds = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error
```

```python
original_score = mse(original_preds, y_test)
```

Now let's drop the outliers and see the effect it has on the score

```python
df.drop(index = df[df.MEDV == 50].index, inplace = True)
```

```python
df.shape
```

Our original dataframe had 506 rows and now we dropped the outliers and have 490.

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size= 0.2, random_state=23)
```

```python
lr.fit(x_train, y_train)
```

```python
new_predictions = lr.predict(x_test)
```

```python
new_score = mse(new_predictions, y_test)
```

```python
plt.bar(['old', 'new'],[original_score, new_score])
plt.title("MSE before and after dropping outliers")
```

# scale transformations

```python
f, ax = plt.subplots(1,2, figsize=(12,6))
ax[0].hist(df.LSTAT);
ax[1].hist(np.log(df.LSTAT)); #converting the LSTAT values to a logarthimic scale using numpy
ax[0].set_title("Before Log Transformation")
ax[1].set_title("After Log Transformation")
```

Notice that after the logarithmic transformation the distribution is more "bell shaped". As most machine learning models aim to generalize based on averages this will allow the model to make better predictions. Typically when the original data is skewed to the right as in this example, a log transformation will "squeeze" the data towards the middle resulting in a "more" normal distribution.

Visualizing log transformations via scatter plot

```python
f, ax = plt.subplots(1, 2, figsize=(10, 6))
ax[0].scatter(df.LSTAT, df.MEDV)
ax[0].set_title("Before Log Transformation")
ax[1].scatter(np.log(df.LSTAT), df.MEDV)
ax[1].set_title("After Log Transformation")
```

In the scatter plot we notice that the log transformation has "squeezed" the far right side of the data resulting in a more linear relationship between LSTAT and MEDV. Let's now see how this can effect our model score.

First convert our feature to Log scale.

```python
df["LSTAT"] = np.log(df["LSTAT"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size= 0.2, random_state=23)
```

```python
lr.fit(x_train, y_train)
```

```python
log_preds = lr.predict(x_test)
```

```python
log_score = mse(log_preds, y_test)
```

```python
plt.barh(['original', 'dropped outliers', 'Outliers/Log Transformation'], [original_score, new_score, log_score])
plt.title("MSE before and after dropping outliers")
```

# We can see that our model is performing significantly better after dropping outliers and performing a logarithmic transformation of only one of the features. Try performing some transformations yourself on more features and see how much you can improve this score.
